* approximation of the distribution using a linear combination of normal distributions

* further reading: https://seaborn.pydata.org/tutorial/distributions.html#kernel-density-estimation

## How do I use a dataframe with seaborn?

We set `data` to be the dataframe, and *name the columns* for `x` and `y`...

```python
sns.scatterplot(data=ti, x='age', y='fare')
```

## How do I create a distribution plot?

```python
sns.distplot(ti['survived'], vertical=True, kde=False)
```

## How do I create a violin plot?

The width of a violin plot is the frequency (ie., how common) a value is within some column...

```python
sns.violinplot(ti['age'])
```

## How do I create a box plot?

Box plots show quarties (25th, 50th, 75th) and outliers.

```python
sns.boxplot(ti['age'])
```

## How do I create a violin plot for multiple columns?

A violin plot can show multiple distributions, each a subset of a single column, factored (or grouped) by another.

```python
sns.violinplot(data=ti, x='survived', y='age')
```

```python
sns.violinplot(data=ti, x="age", y='embark_town')
```

## How do I create a bar plor for multiple columns?

Bar plots are useful for discrete data or for showing, here, the means:

```python
sns.barplot(data=ti, x='survived', y='age');
```

## How do I create a line plot?

```python
sns.lineplot(data=ti, x='fare', y='age');
```

```python
import pandas as pd
ti = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/EY_batch6_2ndSep24/main/titanic.csv')
ti.head(3)
```

```python
import seaborn as sns
```

```python
sns.distplot(ti['survived'], kde=False)
```

```python
sns.distplot(ti['adult_male'], kde=False)
```

```python
for cname in ti.columns:
  print(cname, end=" ")
  print(ti[cname].value_counts().count())
```

```python
rule = lambda val: 'young' if val=='child' else 'old'
ti['age_group'] = ti['who'].apply(rule)
```

```python
ti['age_group'].value_counts()
```

```python
sns.violinplot(data=ti, x='age_group', y='survived')
#old vs. young
```

```python
sns.violinplot(data=ti, x='sex', y='survived')
#men and women
```

```python
rule = lambda val: 'man' if val=='man' else 'woman&child'
ti['isMen'] = ti['who'].apply(rule)
```

```python
ti['isMen'].value_counts()
```

```python
sns.barplot(data=ti, x=ti['isMen'], y=ti['survived'])
#men, women&child
```

```python
sns.violinplot(data=ti, x='who', y='survived')
#Women, child, men
```

```python
rule2 = lambda val: 'First class' if val=='First' else 'Second or Third class'
ti['fair_group'] = ti['class'].apply(rule2)
```

```python
ti['fair_group'].value_counts()
```

```python
sns.barplot(data=ti, x=ti['fair_group'], y=ti['survived'])
```

```python
ti['deck'].value_counts()
```

```python
sns.barplot(data=ti, x=ti['deck'], y=ti['survived'])
```

```python
sns.barplot(data=ti, x=ti['class'], y=ti['survived'])
```

```python
titanic_deck = ti.dropna(subset=['deck'])
titanic_deck
```

```python
surival_chance = titanic_deck.groupby(['deck', 'class'])['survived'].mean()
surival_chance
rule3 = lambda class_val, deck_val: 1 if (class_val == 'First' and deck_val in ['B', 'C', 'D', 'E']) \
                                    else (1 if (class_val == 'Second' and deck_val in ['D', 'E', 'F']) \
                                    else (1 if (class_val == 'Third' and deck_val in ['E', 'F', 'G']) else 0))
titanic_deck['survival_chance'] = titanic_deck.apply(lambda x: rule3(x['class'], x['deck']), axis=1)
titanic_deck['survival_chance']
```

```python
sns.barplot(data=titanic_deck, x=titanic_deck['survival_chance'], y=titanic_deck['class'])
#Survival_chance with deck, class
```
